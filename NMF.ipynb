```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import NMF
```

```python
df=pd.read_csv('articles.csv')
```

```python
df
```

```python
x=df.isnull().sum()
print(x)
```

```python
abc=CountVectorizer(stop_words='english')
```

```python
x=abc.fit_transform(df)
```

```python
topic=5
NMF=NMF(n_components=topic, random_state=42)
NMF.fit(x)
```

```python
words=10
feature= abc.get_feature_names_out()

def display (model,feature,words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic{topic_idx +1 }:")
        print(", ".join([feature[i] for i in topic.argsort ()[:-words - 1:-1]]))

display(NMF,feature,words)
```
